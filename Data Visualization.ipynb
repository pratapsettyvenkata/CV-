```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import matplotlib.image as mpimgi
```

```python
sns.set(style="darkgrid")
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('iris')
```

```python
df.head(10)
```

**Lineplot**

```python
sns.lineplot(x = 'sepal_length',y = 'sepal_width',color='green',data=df)
```

```python
sns.lineplot(x = 'petal_length',y = 'petal_width',color='red',data=df)
```

**scatter plot**

```python
sns.scatterplot(x = 'sepal_length',y = 'sepal_width',color='blue',data=df)
```

```python
sns.scatterplot(x = 'petal_length',y = 'petal_width',color='red',data=df)
```

```python
sns.barplot(x = 'sepal_length',y = 'sepal_width',color='blue',data=df)
```

```python
sns.barplot(x = 'petal_length',y = 'petal_width',color='green',data=df)
```

**Histogram**

```python
sns.histplot(x = 'sepal_length',y = 'sepal_width',color='blue',data=df)
```

```python
sns.histplot(x = 'petal_length',y = 'petal_width',color='green',data=df)
```
